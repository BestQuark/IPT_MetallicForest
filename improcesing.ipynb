```python
vs = np.array([5,7.5, 10])
cu05 = np.array([1.9627, 1.800, 1.703])
cu10 = np.array([1.9629, 1.9532, 1.6651])
zn05 = np.array([1.7229, 1.8579, 1.8337])
zn10 = np.array([1.6924, 1.9302, 1.5627])
```

```python
plt.errorbar(vs, cu05,  yerr=0.05*cu05, marker = 'o', capsize = 5, c='b', label = r'$0.5 M \quad CuSO_4 $')
plt.errorbar(vs, cu10,  yerr=0.05*cu10, marker = 'o', capsize = 5, c='r', label = r'$1.0 M \quad CuSO_4 $')

plt.legend()
plt.xlabel('Voltage [V]')
plt.ylabel('Fractal dimension')
```

```python
plt.errorbar(vs, zn05,  yerr=0.05*zn05, marker = 'o', capsize = 5, c='green', label = r'$0.5 M \quad ZnSO_4 $')
plt.errorbar(vs, zn10,  yerr=0.05*zn10, marker = 'o', capsize = 5, c='purple', label = r'$1.0 M \quad ZnSO_4 $')


plt.legend()
plt.xlabel('Voltage [V]')
plt.ylabel('Fractal dimension')
```
